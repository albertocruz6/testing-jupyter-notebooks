```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
t = np.arange(0,10,1)
v1 = list()
v2 = list()
```

```python
for i in t:
    v1.append(np.exp(i))
    v2.append(2.3**i)
v1 = np.array(v1)
v2 = np.array(v2)
```

```python
df = pd.DataFrame({'t':t, 'v1':v1, 'v2':v2})
```

```python
plt.plot('t','v1',data=df,label="v1")
plt.plot('t','v2',data=df,label="v2")
plt.legend()
plt.show()
```
